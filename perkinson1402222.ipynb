```python
plt.boxplot(df)
plt.title("detect outliers")
plt.xlabel("sample")
```

```python
plt.figure(figsize=(16,15))
sns.heatmap(df.corr(),annot=True,fmt=".2f",cmap= 'RdPu')
plt.show()
```

```python
from scipy import stats
np.abs(stats.zscore(df))
```

```python
df_zscore=df[(np.abs(stats.zscore(df))<3).all(axis=1)]
plt.boxplot(df_zscore)
plt.title("detect outliers")
plt.xlabel("sample")
```

```python
plt.figure(figsize=(16,15))
sns.heatmap(df_zscore.corr(),annot=True,fmt=".2f",cmap= 'RdPu')
plt.show()
```
